```python
import pandas as pd
import matplotlib.pyplot as plt
import requests as req
from census import Census
from us import states
import scipy.stats as stats
import os
import math

# Census API Key
c = Census("5e99f22848e97d5bcc53f03a358b973493473cd9")
cKey = "5e99f22848e97d5bcc53f03a358b973493473cd9"
```

```python
# Read Main CSV
msgPpl = pd.read_csv("NCMEC_final.csv")
msgPpl.head()
```

```python
# get the male vs female missing people pie chart
msngFcnt = msgPpl.groupby(["gender"]).count()["NCMEC_case#"][0]
msngMcnt = msgPpl.groupby(["gender"]).count()["NCMEC_case#"][1]
totMsng = msngFcnt + msngMcnt
msngFper = (msngFcnt / totMsng ) *100
msngMper = (msngMcnt / totMsng)  *100
print (totMsng)
```

```python
# Set Title
title = "Missing People By Gender "

# Labels for the sections of our pie chart
labels = ["Females", "Males"]

# The values of each section of the pie chart
sizes = [msngFper, msngMper]

# The colors of each section of the pie chart
colors = [ "pink", "blue"]

explode = (0.07, 0)

plt.pie(sizes, explode=explode, colors=colors,
        autopct="%1.5f%%", shadow=True, startangle=90, labels=labels)
# plt.legend(title = title, loc="best")
plt.show()
```

```python
stateMsngCnt = msgPpl.pivot_table(index='missing_state', aggfunc={'NCMEC_case#':'count'}).reset_index()
stateMsngCnt = pd.DataFrame(stateMsngCnt)
stateMsngCnt.plot(x=stateMsngCnt['missing_state'],kind='bar',figsize=(22,10),fontsize=14,color='r')
stateMsngCnt = msgPpl.pivot_table(index='missing_state', aggfunc={'NCMEC_case#':'count'}).reset_index()
stateMsngCnt = pd.DataFrame(stateMsngCnt)
stateMsngCnt.plot(x=stateMsngCnt['missing_state'],kind='bar',figsize=(22,10),fontsize=14,color='r')
plt.title("Missing People Count by State")
plt.xlabel("U.S. State")
plt.ylabel("Count  Missing People")
plt.show()
```

```python
EthMsngCnt = msgPpl.pivot_table(index=['ethnicity'], aggfunc={'NCMEC_case#':'count'}).reset_index()
EthMsngCnt = pd.DataFrame(EthMsngCnt)
EthMsngCnt["Ethnicity %"] = round((EthMsngCnt["NCMEC_case#"]/totMsng)*100, 2)
EthMsngCnt = EthMsngCnt[['ethnicity','Ethnicity %']]
EthMsngCnt
```

```python
EthMsngCnt.plot(x=EthMsngCnt['ethnicity'],kind='bar',figsize=(12,10),fontsize=14,color='g')
plt.xlabel("Ethnicity")
plt.ylabel("% Missing People")
title ="Missing persons by Ethnicity"
plt.legend(title = title, loc="upper left")
plt.show()
```

```python
#Create census for states
url = "https://api.census.gov/data/2014/pep/natstprc?get=STNAME,POP&for=state:*&DATE=7"
reqUrl = url + "&key="+ cKey
getData = req.get(reqUrl).json()
df_state_pop = pd.DataFrame(getData,columns=getData[0])
df_state_pop = df_state_pop.drop(df_state_pop.index[0])
df_state_pop['POP'] =  list(map(int,df_state_pop['POP']))
df_state_pop = df_state_pop[df_state_pop.STNAME != 'District of Columbia']
df_state_pop = df_state_pop[df_state_pop.STNAME != 'Puerto Rico Commonwealth']
df_state_pop = df_state_pop.reset_index(drop = True)
msgState = msgPpl.groupby("missing_state").count()["name"]
df_state_pop['Missing_Population'] = list(msgState)
df_state_pop['Missing_Ratio'] = round(df_state_pop['Missing_Population']/df_state_pop['POP'] * 1000000,2)
# df_state_pop.loc[df_state_pop['Missing_Ratio'] > 20]
df_state_pop.set_index(df_state_pop["STNAME"])
```
